```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
URL = "https://www.amazon.in/s?k=playstation5&crid=45AA12CH0P5U&sprefix=playstation%2Caps%2C305&ref=nb_sb_noss_1"
```

```python
# Headers for request
HEADERS = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36', 'Accept-Language': 'en-US, en;q=0.5'})
```

```python
# HTTP Request
webpage = requests.get(URL,headers=HEADERS)
```

```python
webpage
```

```python
# webpage.content
```

```python
# Soup Object containing all data
soup = BeautifulSoup(webpage.content, "html.parser")
```

```python
soup
```

```python
# Fetch links as List of Tag Objects
links = soup.find_all("a", attrs={'class':'a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal'})
```

```python
links
```

```python
link = links[0].get('href')
```

```python
link
```

```python
product_list = "https://amazon.in" + link
```

```python
product_list
```

```python
new_webpage = requests.get(product_list,headers=HEADERS)
```

```python
new_webpage
```

```python
# Soup Object containing all data
new_soup = BeautifulSoup(new_webpage.content,"html.parser")
```

```python
new_soup
```

```python
new_soup.find("span",attrs={"id":'productTitle'}).text.strip()
```

```python
new_soup.find("span",attrs={"class":'a-price-whole'}).text.strip()
```
